```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# dataset
x = np.linspace(0, 10, 100)
y = np.sin(x)

# plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, linestyle='--', color='blue')

# title and labels
plt.title('Sin Plot')
plt.xlabel('X')
plt.ylabel('sin(X)')
plt.legend()

# plot
plt.show()
```

```python
import seaborn as sns

# tips dataset
df = sns.load_dataset('tips')

# figure
plt.figure(figsize=(5, 4))

# Task 1: histogram
sns.histplot(df['total_bill'], bins=20, kde=True, color='red')

# title and labels
plt.title('Bill Distribution')
plt.xlabel('Bill Total')
plt.ylabel('Bill Frequency')

# plot
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Box plot
sns.boxplot(x='time', y='tip', hue= 'smoker', data=df, ax=axes[0], palette='dark')
axes[0].set_title('Box Plot of Tips')
axes[0].set_xlabel('Tip Time')
axes[0].set_ylabel('Tip Amount')

# Violin plot
sns.violinplot(x='time', y='tip', hue= 'smoker', data=df, ax=axes[1], palette='dark')
axes[1].set_title('Violin Plot of Tips')
axes[1].set_xlabel('Tip Time')
axes[1].set_ylabel('Tip Amount')

# plot
plt.show()
```

```python
# Create scatter plot with regression lines
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=df, palette='dark', aspect=1.5)

# title and labels
plt.title('Bill and Tip Smoker Association')
plt.xlabel('Total Bill')
plt.ylabel('Total Tip')

# plot
plt.show()
```

```python
# correlation matrix with numbers
corr_matrix = df.select_dtypes(include=['number']).corr()

# heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

#title
plt.title('Correlation Matrix of Tips ')

# plot
plt.show()
```
